```python
def plotaudio(file,sampleRate):
    sig , sr = librosa.load(file,mono=True,sr=sampleRate)
    time=np.arange(0,sig.shape[0])/sr
    excursion=max(abs(sig))
    sigNew = sig / excursion
   # Label the x and y axes with variables
    xlabel = 'Time (seconds)'
    ylabel = 'Amplitude'
    # plot in soundwave
    plt.plot(time,sigNew)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
plotaudio('avm.wav',44100)
```

```python
plotaudio('avm.wav',441)
```
